```python
import random
secrate_no = random.randint(1,100)
attempts = 0;

while True:
  guess = int(input("Guess a number between 1 to 100"))
  attempts += 1

  if guess <secrate_no:
    print("Too low!")
  elif guess> secrate_no:
    print("Too high!")
  else:
    print("correct attempts", attempts)
    break
```
