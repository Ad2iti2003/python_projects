To-Do App

```python
tasks = {}

while True:
  print("\n1. Add Task")
  print("2. View Task")
  print("3. Mark Complete")
  print("4. Delete Task")
  print("5. exit")

  choice = input("Enter choice")

  if choice == '1':
    task = input("Enter task: ")
    status = input("Enter status: ")
    tasks[task] = status

  elif choice == '2':
    for task , status in tasks.items():
      print(task, "-", status)

  elif choice == '3':
    task = input("Enter task mark to complete: ")
    if task in tasks:
      tasks[task] = "completed"

  elif choice == '4':
    task = input("Enter task to be deleted")
    tasks.pop(task,None)

  else:
    break
```
